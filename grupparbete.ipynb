```python
# ---------- Subscriber ----------
# Connect to HiveMQ Cloud and subscribe to your topic
# Note that MQTT is a real-time publish/subscribe-protocol

import paho.mqtt.client as mqtt
import json
import csv
import os

# MQTT Broker
broker =  "0fbfc1fe1c1b4bb0b4465a9d0ad66dd7.s1.eu.hivemq.cloud"
port = 8883
username = "grupparbete"
password = "Marvel12"
topic = "room1/environment"

# CSV-file
filename = "iot_data.csv"

if not os.path.exists(filename):
    with open(filename, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["timestamp", "device_id", "temperature", "humidity", "alert"])
```

```python
def on_message(client, userdata, msg):
    data = json.loads(msg.payload.decode())

    print("Received:", data)

    # Skriv till CSV
    with open(filename, mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([
            data["timestamp"],
            data["device_id"],
            data["temperature"],
            data["humidity"],
            data["alert"]
        ])
```

```python

# MQTT callback
def on_connect(client, userdata, flags, rc):
    print("Connected with result code", rc)
    client.subscribe(topic)

```

```python

# MQTT client
client = mqtt.Client()
client.username_pw_set(username, password)
client.tls_set()

client.on_connect = on_connect
client.on_message = on_message

client.connect(broker, port)

print("Listening for data...")

client.loop_forever()


```
